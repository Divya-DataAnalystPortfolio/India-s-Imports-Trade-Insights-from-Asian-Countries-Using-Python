**5. Defining Valid Unit Values**

```python
# Since unit of quantity has inconsistent and null values, defining a Valid unit values for Unit of Quantity column
valid_units = ['Kgs','Nos','Sqm','Prs','Cbm','Ctm','Ltr','Tkw','Mts','Ton','Kg','C/K','Mtr','Thd']
```

**6. Cleaning the Column by Replacing Invalid Values**

```python
# Converting invalid values with NaN
df['Unit of Quantity (unit)'] = df['Unit of Quantity (unit)'].apply(lambda x: x if x in valid_units else np.nan)
```

**7. Verifying the Cleaning Process**

```python
# Display the no. of unique values in a column
df['Unit of Quantity (unit)'].nunique()
```

```python
# Count occurences of each unique value in a column
df['Unit of Quantity (unit)'].value_counts(dropna=False)   # Dropna=False will also list the count no. of NaN values
```

**8. a. Handling Missing Values for categorical column using Imputation**

```python
# Fillna: Mode for categorical column
df['Unit of Quantity (unit)'].fillna(df['Unit of Quantity (unit)'].mode()[0],inplace=True)
```

**8. b. Handling Missing Values for numerical column using Imputation**

```python
# Impute missing values for numerical columns using median
num_cols = ['Quantity of commodity (value_qt)',
            'Value of commodity quantity in INR (value_rs)',
            'Value of commodity quantity in US Dollars (value_dl)']

df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

# **3. Mutlivariate Analysis**

```python
# Scatter Plot - Quantity vs Import Value by Unit

plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x='Quantity of commodity (value_qt)',
    y='Value of commodity quantity in US Dollars (value_dl)',
    hue='Unit of Quantity (unit)',
    alpha=0.5
)

plt.title("Quantity vs Import Value by Unit of Quantity")
plt.xlabel("Quantity")
plt.ylabel("Import Value (USD)")
plt.show()
```

**Interpretation:**

This scatter plot shows the relationship between quantity and import value while distinguishing measurement units using colors.

```python
# Stacked Bar Chart - Import Value by Country and Commodity

top_countries = df['Country Name (country_name)'].value_counts().head(5).index

filtered_df = df[df['Country Name (country_name)'].isin(top_countries)]

pivot_table = filtered_df.pivot_table(
    values='Value of commodity quantity in US Dollars (value_dl)',
    index='Country Name (country_name)',
    columns='Commodity Name (commodity)',
    aggfunc='sum'
).fillna(0)

pivot_table.iloc[:, :5].plot(kind='bar', stacked=True, figsize=(10,6))

plt.title("Import Value by Country and Commodity")
plt.xlabel("Country")
plt.ylabel("Import Value (USD)")
plt.show()
```

**Interpretation:**

This chart compares import values across countries and commodities simultaneously, helping identify which commodities dominate imports from specific countries.
